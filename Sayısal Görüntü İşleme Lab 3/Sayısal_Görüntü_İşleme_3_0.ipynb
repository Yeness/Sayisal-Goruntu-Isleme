# Kenar Tespiti - Edge Detection

Türev ile ani renk değişimi

```python
sobel_x = cv2.Sobel(resim, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(resim, cv2.CV_64F, 0, 1, ksize=3)
```

CV_64F = 64 bit float

```python
sobel_x = np.absolute(sobel_x).astype(np.uint8)
sobel_y = np.absolute(sobel_y).astype(np.uint8)
```

```python
sobel_birlestirilmis = cv2.bitwise_or(sobel_x, sobel_y)
```

```python
sobel_birlestirilmis
```

```python
canny_kenar = cv2.Canny(resim, 50, 150)
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(1, 5, 1)
plt.imshow(resim, cmap='gray')
plt.axis('off')

plt.subplot(1, 5, 2)
plt.imshow(sobel_x, cmap='gray')
plt.axis('off')

plt.subplot(1, 5, 3)
plt.imshow(sobel_y, cmap='gray')
plt.axis('off')

plt.subplot(1, 5, 4)
plt.imshow(sobel_birlestirilmis, cmap='gray')
plt.axis('off')

plt.subplot(1, 5, 5)
plt.imshow(canny_kenar, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

#Morfolojik İşlemler

```python
maske = np.zeros((300, 300), dtype=np.uint8)
```

```python
maske
```

```python
maske[100:200, 100:200] = 255
```

```python
maske
```

```python
maske[130:150, 130:150] = 0
maske[170:180, 170:180] = 0
```

```python
maske
```

```python
maske[20:40, 20:40] = 255
maske[250:260, 250:260] = 255
maske[50:60, 200:220] = 255
```

```python
maske
```

```python
kernel = np.ones((9, 9), np.uint8)
```

```python
kernel
```

```python
asinmis = cv2.erode(maske, kernel, iterations=1)        #Erosion
genislemis = cv2.dilate(maske, kernel, iterations=1)    #Dilation
```

```python
acilis = cv2.morphologyEx(maske, cv2.MORPH_OPEN, kernel) #Aşınma - Genişleme
```

```python
kapanis = cv2.morphologyEx(maske, cv2.MORPH_CLOSE, kernel) #Genişleme - Aşınma
```

```python
plt.figure(figsize=(18, 10))

plt.subplot(2, 3, 1)
plt.imshow(maske, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(asinmis, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(genislemis, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(acilis, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(kapanis, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

#HSV

Hue, Saturation, Value - Renk Tonu, Doygunluk, Değer

Hue - Nesnenin hangi renk olduğunu belirleyen kod

Saturation - O rengin ne kadar canlı olduğunu belirler

Value - Ortamdaki ışık miktarını belirler

```python
resim_rgb = np.ones((200, 500, 3), dtype=np.uint8) * 255
```

```python
resim_rgb
```

```python
resim_rgb[50:150, 50:150] = [0, 0, 255]
resim_rgb[50:150, 200:300] = [0, 0, 80]
resim_rgb[50:150, 350:450] = [255, 0, 0]
```

```python
resim_rgb
```

```python
resim_hsv = cv2.cvtColor(resim_rgb, cv2.COLOR_RGB2HSV)
```

```python
resim_hsv
```

```python
alt_mavi = np.array([100, 50, 50])
ust_mavi = np.array([140, 255, 255])
```

```python
mavi_maske = cv2.inRange(resim_hsv, alt_mavi, ust_mavi)
```

```python
sadece_maviler = cv2.bitwise_and(resim_rgb, resim_rgb, mask=mavi_maske)
```

```python
plt.figure(figsize=(15, 8))

plt.subplot(2, 2, 1)
plt.imshow(resim_rgb)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(resim_hsv)
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(mavi_maske, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(sadece_maviler)
plt.axis('off')

plt.tight_layout()
plt.show()
```

#Kontur Bulma ve Çerçeveleme

```python
maske = np.zeros((300, 400), dtype=np.uint8)
```

```python
maske
```

```python
maske[50:120, 50:120] = 255
maske[150:250, 200:350] = 255
maske[200:220, 250:270] = 0
maske[20:30, 300:310] = 255
```

```python
maske
```

```python
sonuc_resmi = cv2.cvtColor(maske, cv2.COLOR_GRAY2RGB)
```

```python
sonuc_resmi
```

```python
konturlar, hiyerarsi = cv2.findContours(maske, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

RETR_EXTERNAL -> sadece dış sınırları bul
CHAIN_APPROX_SIMPLE -> noktaları hafızada tut - köşe olursa daha iyi

```python
len(konturlar)
```

```python
for kontur in konturlar:
    alan = cv2.contourArea(kontur)

    if alan > 100:
        x, y, w, h = cv2.boundingRect(kontur)

        cv2.rectangle(sonuc_resmi, (x, y), (x + w, y + h), (0, 255, 0), 2)

        cv2.putText(sonuc_resmi, f"Alan:{alan}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(maske, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sonuc_resmi)
plt.axis('off')

plt.tight_layout()
plt.show()
```

#Histogram Germe ve Eşitleme

```python
matris = np.array([
    [2, 3, 3, 2],
    [4, 2, 4, 3],
    [5, 4, 2, 3],
    [2, 3, 5, 4]
], dtype=np.uint8)
```

```python
matris
```

Amaç resmi daha canlı hale getirmektir

```python
min_val = matris.min()
max_val = matris.max()
L_minus_1 = 7
```

Yeni Değer = round((Eski Değer - Min)/(Max-Min) * 7)

```python
germis_matris = np.round((matris - min_val) / (max_val - min_val) * L_minus_1).astype(np.uint8)
```

Siyahi siyah beyazı beyaz yapar

```python
germis_matris
```

görüntüyü bozarak saklı olanı çıkarır

```python
hist, _ = np.histogram(matris.flatten(), bins=8, range=[0, 8])  # etiketleri almadık
```

çok olanın yeri çok, az olanın yeri az

```python
hist
```

```python
cdf = hist.cumsum()
```

```python
cdf
```

```python
toplam_piksel = matris.size
cdf_normalize = np.round((cdf * L_minus_1) / toplam_piksel).astype(np.uint8)
```

```python
esitlenmis_matris = cdf_normalize[matris]
```

```python
esitlenmis_matris
```
